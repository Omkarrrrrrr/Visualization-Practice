```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.random.rand(50)
y= np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c = '#223BDC' ,alpha=.9,marker='X')
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
plt.grid()
```

```python
x = np.linspace(1,10 , 100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y , '--r')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sin wave graph")
plt.grid()
```

```python
x = ['a', 'b' , 'c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color = '#7E10D5')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color = '#7E10D5')
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```
